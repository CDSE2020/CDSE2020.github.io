<div class="bb alert alert-info">
<h2 class="bbh">
Lists Comprehensions
</h2>

<hr/>

<ul>
    <li>Creates a new list from a given list.</li>
</ul>

<pre class = "bb" style="width:340px">
<code class>[f(x) for x in mylist]
[f(x) for x in mylist if condition]
</code></pre>

</div>

```python
numbers = list(range(10))
numbers
```

```python
squares = [x**2 for x in numbers]
squares
```

```python
even_squares = [x**2 for x in numbers if x%2 == 0]
even_squares
```

The same with the "for" loop:

```python
even_squares2 = []
for x in numbers:
    if x%2 == 0:
        even_squares2.append(x**2)

even_squares2
```

```python
from time import sleep

def timer(seconds, message = "Time's up!"):
    for t in range(seconds,0,-1):
        print(t)
        sleep(1)
    print(message)
```

```python
timer(10)
```

```python
timer(5, message = "Boom!")
```

```python
with open("conrad.txt") as foo:
    s = foo.read()
    
print(s)
```

```python
len(s)
```

```python
slow = s.lower()
print(slow)
```

```python
chars = set(slow)
print(chars)
```

```python
import string
string.ascii_lowercase
```

```python
remove = chars - set(string.ascii_letters + " ")
print(remove)
```

```python
for c in remove:
    slow = slow.replace(c, " ")
print(slow)
```

```python
words = slow.split()
```

```python
print(words)
```

```python
len(words)
```

```python
len(set(words))
```

```python
word_dict = {}
for w in words:
    if w in word_dict:
        word_dict[w] += 1
    else:
        word_dict[w] = 1
```
