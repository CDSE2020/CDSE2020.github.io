```python
from bs4 import BeautifulSoup
```

<div class="bb alert alert-info">
Parse the page:
</div>

```python
soup = BeautifulSoup(grads)
```

<div class="bb alert alert-info">
Print the page code neatly:
</div>

```python
print(soup.prettify())
```

<div class="bb alert alert-info">
Find all HTML <code>&lt;p&gt;...&lt;/p&gt;</code> tags:
</div>

```python
ptags = soup.find_all('p')
ptags
```

<div class="bb alert alert-info">

<p>The <code>get_text()</code> method returns text enclosed in a tag.</p>
</div>

```python
ptags[10]
```

```python
ptags[10].get_text()
```

```python
ptags[0]
```

<div class="bb alert alert-info">
Get text from all paragraphs containing the word "Office":
</div>

```python
grad_list = [tag.get_text() for tag in ptags if "Office:" in tag.get_text()]
grad_list
```

<div class="bb alert alert-info">
<h1 class="bbh">
String Processing
</h1>
    
<hr/>

Here we get names and emails from the strings.

</div>

```python
s = grad_list[10]
s
```

```python
n = s.find("Office")
n
```

```python
s[0:n].strip()
```

```python
def get_name(s):
    stop = s.find("Office:")
    return s[:stop].strip()
```

```python
get_name(s)
```

```python
start = s.find("Email:")
s[start:].split()
```

```python
def get_email(s):
    start = s.find("Email:")
    return s[start:].split()[1]

get_email(s)
```

```python
for s in grad_list:
    name = get_name(s)
    email = get_email(s)
    print(f"Name:   {name}")
    print(f"Email:  {email}")
    print()
```
